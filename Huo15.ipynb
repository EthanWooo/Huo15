```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1=np.array(Image.open('./data/img_001.png'))
img2=np.array(Image.open('./data/img_002.png'))
img3=np.array(Image.open('./data/img_003.png'))
img4=np.array(Image.open('./data/img_004.png'))
img5=np.array(Image.open('./data/img_005.png'))
img6=np.array(Image.open('./data/img_006.png'))
img7=np.array(Image.open('./data/img_007.png'))
img8=np.array(Image.open('./data/img_008.jpg'))
# print(img.shape)
# print(img.size)
# print (img)
```

```python
def getGray(img):
    return img[:,:,0]*0.33+img[:,:,1]*0.33+img[:,:,2]*0.33

def highlightDistinguish(img):
    N = img.shape[0]*img.shape[1]
    SF = img.copy()
    MSF = img.copy()
    plainArea = img.copy()
    highlightArea = img.copy()
    temp_sum = 0
    
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            temp_sum = temp_sum+img[x,y,:].min()
            
    I_min = np.array([temp_sum/N,temp_sum/N,temp_sum/N])
    
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            pixel_min = np.array([img[x,y,:].min(),img[x,y,:].min(),img[x,y,:].min()])
            SF[x,y,:] = img[x,y,:] - pixel_min
            MSF[x,y,:] = SF[x,y,:] + I_min

    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            if(((img[x,y,:]-I_min)>I_min).all()):
                plainArea[x,y,:]=[0,0,0]
            else:
                highlightArea[x,y,:]=[0,0,0]
    return highlightArea, plainArea,MSF

def getHDR(img):
    N = img.shape[0]*img.shape[1]
    L_highlight,L_plain,MSF = highlightDistinguish(img)
    sum_r = 0
    sum_g = 0
    sum_b = 0
    for x in range(img.shape[0]):
        for y in range(img.shape[1]):
            sum_r = sum_r+(np.log(MSF[x,y,0].astype(np.float)+1/256))/N
#             print(sum_r)
            sum_g = sum_g+(np.log(MSF[x,y,1].astype(np.float)+1/256))/N
            sum_b = sum_b+(np.log(MSF[x,y,2].astype(np.float)+1/256))/N
    average_luminance_value = np.array([np.exp(sum_r),np.exp(sum_g),np.exp(sum_b)])
    
    return average_luminance_value
```

```python
highlightPart1,plainPart1,_ = highlightDistinguish(img1)
im = Image.fromarray(plainPart1)
im.save('./output/out_001.png')

plt.imshow(img1)
plt.axis('off')
plt.show()
plt.imshow(highlightPart1)
plt.axis('off')
plt.show()
plt.imshow(plainPart1)
plt.axis('off')
plt.show()
```
